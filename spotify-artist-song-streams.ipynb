```python
file="../input/spotifys-worldwide-daily-song-ranking/data.csv"
df= pd.read_csv(file, index_col="Position")
df['Date']= pd.to_datetime(df['Date'])
df
```

```python
artists=df['Artist'].unique()
df=df.sort_values('Artist').dropna()
df
```

```python
artist= input("Enter artist: ")
```

```python
df1=df.where(df['Artist']==artist).dropna().sort_values('Track Name')
df1# artist's songs
```

```python
df2=df1.drop(['URL','Region','Artist'],axis=1)
df2# still has duplicate dates
```

```python
df2=df2.reset_index()
df2
```

```python
df3=df2[['Track Name','Streams']].copy()
df3
```

```python
#df3.where(df3['Track Name']=="Afire Love").dropna()
```

```python
grp=df3.groupby(['Track Name']).sum()
df4=grp.apply(lambda a: a[:])
df4
```

```python
plt.figure(figsize=(20,10))
g=sns.barplot(x=df4.index, y=df4['Streams'])
g.set_xticklabels(df4.index,rotation=90)
plt.title('songs streamed')
plt.legend()
print("Most streamed song:")
print(df4['Streams'].idxmax())
print("")
print("Least streamed song:")
print(df4['Streams'].idxmin())
```
